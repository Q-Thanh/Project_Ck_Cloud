### tiếp theo chúng ta sẽ viết code giải nén bộ data vào bộ nhớ tạm của máy ảo để quá trình huấn luyện nhanh hơn khi đọc trực tiếp từ volume

```python
import zipfile
import os

# Đường dẫn chính xác từ Volume của bạn
zip_path = '/Volumes/workspace/default/dataset_ổ_gà/pothole.v8i.yolov8.zip'
# Thư mục chứa dữ liệu sau khi giải nén
extract_path = '/tmp/dataset_mat_than'

print("Đang tiến hành giải nén, vui lòng đợi...")

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print(f"Giải nén thành công vào: {extract_path}")
print("Các file bên trong:", os.listdir(extract_path))
```

### tiếp tục chúng ta sẽ bắt đầu huấn luyện và tự đặt epochs = 5 để chạy cho nhanh và chỉ học 1 lần 2 ảnh cho nhẹ ram

```python
from ultralytics import YOLO

# Tải mô hình YOLOv8 nano
model = YOLO('yolov8n.pt') 

yaml_path = '/tmp/dataset_mat_than/data.yaml'

print("BẮT ĐẦU DẠY AI (CHẾ ĐỘ SIÊU NHẸ)...")
results = model.train(
    data=yaml_path, 
    epochs=5, 
    imgsz=320,       # Thu nhỏ ảnh lại để cực kỳ tiết kiệm RAM
    batch=2,         # Nhồi từng 2 ảnh một
    workers=0,       
    device='cpu',    # BẮT BUỘC: Ép chạy CPU để không bị treo khi tìm GPU
    cache=False      # Cấm lưu đệm vào RAM
)
```
